```python
import requests
import csv
import pandas as pd
import json
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
import time
import matplotlib.pyplot as plt

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

with open('iex_token.txt') as f: 
    api_key = f.read()

symbol = 'AAPL'


api_url = f'https://cloud.iexapis.com/stable/stock/{symbol}/chart/max?token={api_key}'

df = pd.read_json(api_url)

#initiate the application
app = JupyterDash(__name__, external_stylesheets = external_stylesheets)

#define the layout
app.layout = html.Div(
    id = 'app-container',
    children = [
        html.Div(
            id = 'header-container',
        children = [html.H2("Historical Stock Price of Apple")]),
        html.Div(
            id = 'body-container',
            children = [
                dcc.Graph(id = 'close-price',
                          style = {'height' : '600px'},
                         figure = {'data' : [{'x' : df.query("symbol == 'AAPL'").date,
                                             'y' : df.query("symbol == 'AAPL'").close,
                                             'type' : 'line'}],
                                  '           layout' : {'title' : 'Historical Stock Price of Apple'}}),
                                                ])
                ])
    


#app.run_server(mode='inline', debug = False)

if __name__ == '__main__':
        app.run_server(mode='inline', debug=False, port=8051)
```
